```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, n_classes=2, random_state=3)
print(X.shape, y.shape)

plt.grid(True)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr')
```

```python
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

logisticRegression:

```python
model1 = LogisticRegression(solver='saga', max_iter=20, random_state=3)
model1.fit(x_train, y_train)
model1.predict(x_test), y_test
```

```python
model1.score(x_train, y_train)
```

```python
model1.score(x_test, y_test)
```

SGDClassifier:

```python
model2 = SGDClassifier(loss='log_loss', max_iter=10, learning_rate='constant', eta0=0.2, random_state=3)
model2.fit(x_train, y_train)
model2.predict(x_test), y_test
```

```python
model2.score(x_train, y_train)
```

```python
model2.score(x_test, y_test)
```

```python
X = df_shuffled[['Feature1', 'Feature2', 'Feature3', 'Feature4']].values
y = df_shuffled['Class'].values
X, y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

part3:

```python
x_train = np.hstack((np.ones((len(x_train), 1)), x_train))
x_train.shape
```

```python
def sigmoid(x):
  return 1 / (1+np.exp(-x))

def logistic_regression(x, w):
  y_hat = sigmoid(x @ w)
  return y_hat

def bce(y, y_hat):
  loss = -(np.mean(y*np.log(y_hat) + (1-y)*np.log(1-y_hat)))
  return loss

def gradient(x, y, y_hat):
  grads = (x.T @ (y_hat - y)) / len(y)
  return grads

def  gradiant_descent(w, eta, grads):
  w -= eta*grads.mean(axis=1, keepdims=True)
  return w

def accuracy(y, y_hat):
  acc = np.mean((y == 1) == (np.round(y_hat)))
  return acc
```

```python
w = np.random.RandomState(3).randn(5, 1)
print(w)

eta = 0.02
n_epochs = 2000
```

```python
error_hist = []

for epoch in range(n_epochs):
  y_hat = logistic_regression(x_train, w)

  e = bce(y_train, y_hat)
  error_hist.append(e)

  grads = gradient(x_train, y_train, y_hat)

  w = gradiant_descent(w, eta, grads)

  if (epoch+1) % 100 == 0:
    print(f'epoch={epoch},  E={e:.3},  w={w.T[0]}')
```

```python
plt.plot(error_hist)
```

```python
accuracy(y_train, y_hat)
```

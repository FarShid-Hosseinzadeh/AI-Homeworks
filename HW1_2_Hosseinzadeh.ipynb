#Problem10

```python
import pandas as pd

data = {
    'Student ID': [1, 2, 3, 4, 5],
    'Name': ['Farbod', 'FarShid', 'Sajjad', 'Dorsa', 'Baran'],
    'Score': [88, 92, 76, 65, 84]
}

df = pd.DataFrame(data)

def determine_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    else:
        return 'D'

df['Grade'] = df['Score'].apply(determine_grade)

print(df)
```

#Problem11

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {
    'Month': pd.date_range(start='2021-01-01', periods=24, freq='M'),
    'Product A Sales': [500, 480, 600, 750, 900, 850, 920, 1100, 1300, 1350, 1500, 1450, 1550, 1600, 1650, 1600, 1500, 1400, 1600, 1700, 1800, 1750, 1850, 1900],
    'Product B Sales': [300, 320, 400, 450, 500, 580, 700, 750, 820, 900, 950, 980, 1050, 1100, 1150, 1200, 1250, 1300, 1350, 1400, 1500, 1600, 1550, 1700],
    'Product C Sales': [200, 210, 250, 280, 320, 350, 380, 400, 420, 440, 460, 480, 500, 520, 540, 560, 580, 600, 620, 640, 660, 680, 700, 720]
}

sales_df = pd.DataFrame(data)

print(sales_df.head(2),'\n')
print(sales_df.tail(2))

total_sales = sales_df.sum(numeric_only=True)
print('\n',total_sales,'\n')

max_sales_month_A = sales_df[sales_df['Product A Sales'] == sales_df['Product A Sales'].max()]['Month'].values[0]
max_sales_month_B = sales_df[sales_df['Product B Sales'] == sales_df['Product B Sales'].max()]['Month'].values[0]
max_sales_month_C = sales_df[sales_df['Product C Sales'] == sales_df['Product C Sales'].max()]['Month'].values[0]
print(f"Product A had its highest sales in {max_sales_month_A}")
print(f"Product B had its highest sales in {max_sales_month_B}")
print(f"Product C had its highest sales in {max_sales_month_C}",'\n')

min_sales_month_A = sales_df[sales_df['Product A Sales'] == sales_df['Product A Sales'].min()]['Month'].values[0]
min_sales_month_B = sales_df[sales_df['Product B Sales'] == sales_df['Product B Sales'].min()]['Month'].values[0]
min_sales_month_C = sales_df[sales_df['Product C Sales'] == sales_df['Product C Sales'].min()]['Month'].values[0]
print(f"Product A had its lowest sales in {min_sales_month_A}")
print(f"Product B had its lowest sales in {min_sales_month_B}")
print(f"Product C had its lowest sales in {min_sales_month_C}",'\n')

plt.figure(figsize=(12, 6))
plt.plot(sales_df['Month'], sales_df['Product A Sales'], label='Product A')
plt.plot(sales_df['Month'], sales_df['Product B Sales'], label='Product B')
plt.plot(sales_df['Month'], sales_df['Product C Sales'], label='Product C')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.legend()
plt.grid(True)
plt.show()

print('\n',sales_df.describe())
```
